```python

chile_mont = os.path.join( 'Comuna_Maitencillo.geojson')

with open(chile_mont) as f:
    states = json.load(f)

kw = {'location': [-32.6689, -71.3960]}

m = folium.Map( **kw,attr=None,tiles = None, zoom_start=13,control_scale=True,prefer_canvas=True,detect_retina = True)
#marca lat lot del punto o lugar
#mapa base de thunderforest
folium.raster_layers.TileLayer(
    tiles='https://a.tile.opentopomap.org/{z}/{x}/{y}.png',
    attr= attr,name='Mapa Base OSM',max_zoom=21,subdomains=['mt0', 'mt1', 'mt2', 'mt3'],overlay=False,control=True,).add_to(m)
#tesela base de de google
folium.raster_layers.TileLayer(
    tiles='http://{s}.google.com/vt/lyrs=s&x={x}&y={y}&z={z}',attr= attr2,name='Google Earth',
    max_zoom=21,subdomains=['mt0', 'mt1', 'mt2', 'mt3'],overlay=False,control=True,).add_to(m)
#plugins.ScrollZoomToggler().add_to(m)
plugins.Fullscreen(
    position='topright',title='Expandir',title_cancel='salir',force_separate_button=True).add_to(m)
GeoJson(states).add_to(m)

#controlador de capas 
m.add_child(folium.LayerControl())
#marca lat lot del punto o lugar
m.add_child(folium.LatLngPopup())
Draw(export=None).add_to(m)
plugins.LocateControl().add_to(m)
#mini mapa y un zoom por -
minimap = MiniMap(toggle_display=True,zoom_level_offset=-2)
minimap.add_to(m)
m.save(os.path.join('maitencillo.html'))
m
```

```python
url = 'https://raw.githubusercontent.com/camiloceacarvajal/Visualizacion-folium-CGR/master/productos/vacas.csv'
cementerio_data = pd.read_csv(url,index_col=0,parse_dates=[0])
cementerio_data.describe()
```

```python
#definir de donde se extrae la info 
df = pd.read_csv(url,index_col=0,parse_dates=[0])
#definimos el nombre del mapa y sus caracteristicas iniciales tales como:escala,zoom,coordenadas inicial
m = folium.plugins.DualMap(cementerio_data[['Latitudes','Longitudes']].mean().tolist(),
               attr=None,tiles = None, zoom_start=10,control_scale=True,prefer_canvas=True,detect_retina = True)
tooltip = 'Click me!'
#mapa base de thunderforest
folium.raster_layers.TileLayer(
    tiles='https://tile.thunderforest.com/transport/{z}/{x}/{y}.png?apikey=3cd85f11f4744c0c8c3bdaab8483cde0',
    attr= attr,name='Mapa Base OSM',max_zoom=21,subdomains=['mt0', 'mt1', 'mt2', 'mt3'],overlay=False,control=True,).add_to(m)
#tesela base de de google
folium.raster_layers.TileLayer(
    tiles='http://{s}.google.com/vt/lyrs=s&x={x}&y={y}&z={z}',attr= attr2,name='Google Earth',
    max_zoom=21,subdomains=['mt0', 'mt1', 'mt2', 'mt3'],overlay=False,control=True,).add_to(m)
#plugins.ScrollZoomToggler().add_to(m)
plugins.Fullscreen(
    position='topright',title='Expandir',title_cancel='salir',force_separate_button=True).add_to(m)
# definicion y leer de tablas de atributos 
mc = MarkerCluster(name="RM")
#ciclo desde para generar marcadores con una visual como excel 
for N,data in cementerio_data.iterrows():
    station_html = folium.Html('<b>%s</b>' %(pd.DataFrame(data).to_html()),script=True)
    mc.add_child(folium.Marker([data.Latitudes,data.Longitudes],
 popup = folium.Popup(station_html),tooltip=tooltip,icon=folium.Icon(color='green'if data.provincia == 'Santiago'
                  else 'red',prefix = 'glyphicon',icon='cloud')                              
                    ))

#marcador que dirije a la base de datos y la puedes descargar
ref_html = r'<a href="https://drive.google.com/file/d/1NbJ0huzkKYjsfdTOlN8xT7VExEwd75Vq/view?usp=sharing" target="_blank"> Base de datos </a>'
pop_html = folium.Popup(folium.Html(ref_html, script=True))
folium.Marker([-33.4177, -70.6504],popup = pop_html,tooltip=tooltip,icon=folium.Icon(color='red',icon='clouw')).add_to(m)
#define la variable y el archico json
chile_mont = os.path.join( 'Comuna.json')

with open(chile_mont) as f:
    states = json.load(f)

GeoJson(states).add_to(m)
#controlador de capas 
m.add_child(mc)
m.add_child(folium.LayerControl())
#marca lat lot del punto o lugar
m.add_child(folium.LatLngPopup())
Draw(export=None).add_to(m)
plugins.LocateControl().add_to(m)
#mini mapa y un zoom por -2 
minimap = MiniMap(toggle_display=True,zoom_level_offset=-2)
minimap.add_to(m)
#guardar archivo con un nombre 
m.save( 'Rm1.html')
#m.save(os.path.join('results', 'rm.html'))
m
```
